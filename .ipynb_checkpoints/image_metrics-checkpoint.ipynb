```python
import numpy as np
import skimage
from skimage import measure
```

```python
def mse(img1, img2):
    return np.mean(np.square(img1 - img2))
```

```python
def psnr(img1, img2):
    error = mse(img1, img2)
    if error == 0:
        return 100
    PIXEL_MAX = max(np.amax(img1), np.amax(img2))
    return 20 * np.log10(PIXEL_MAX / np.sqrt(error))
```

```python
def ssim(img1, img2):
    return measure.compare_ssim(img1, img2, multichannel = True)
```

```python
def mae(img1, img2):
    return np.mean(np.abs(img1- img2))
```

```python
def metrics(img1, img2):
#     assert(img1.shape[1:] == (32, 32, 3))
#     assert(img2.shape[1:] == (32, 32, 3))
    psnr_val = psnr(img1, img2)
    ssim_val = ssim(img1, img2)
    mse_val = mse(img1, img2)
    mae_val = mae(img1, img2)
    return psnr_val, ssim_val, mse_val, mae_val
```

```python
import glob

for fn in glob.glob("./outs/*.npz"):

    dae = np.load(fn)
    original = dae['original']
    # encoded = dae['encoded']
    decoded = dae['decoded']

    psnr_val, ssim_val, mse_val, mae_val = metrics(original, decoded)

    print( fn.split("/")[-1].split(".")[0] , "->" , psnr_val , ssim_val , mse_val , mae_val )
```
